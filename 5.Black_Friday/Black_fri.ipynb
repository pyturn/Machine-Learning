```python
Product_Category_2_mean = df['Product_Category_2'].groupby(df['Age']).mean()
Product_Category_2_mean
#Product_Category_2_mean.loc['0-17']
```

```python
#https://stackoverflow.com/questions/13295735/how-can-i-replace-all-the-nan-values-with-zeros-in-a-column-of-a-pandas-datafra
df.loc[(df['Age']=='0-17'),'Product_Category_2']=df.loc[(df['Age']=='0-17'),'Product_Category_2'].fillna(int(Product_Category_2_mean.loc['0-17']))
df.loc[(df['Age']=='18-25'),'Product_Category_2']=df.loc[(df['Age']=='18-25'),'Product_Category_2'].fillna(int(Product_Category_2_mean.loc['18-25']))
df.loc[(df['Age']=='26-35'),'Product_Category_2']=df.loc[(df['Age']=='26-35'),'Product_Category_2'].fillna(int(Product_Category_2_mean.loc['26-35']))
df.loc[(df['Age']=='36-45'),'Product_Category_2']=df.loc[(df['Age']=='36-45'),'Product_Category_2'].fillna(int(Product_Category_2_mean.loc['36-45']))
df.loc[(df['Age']=='46-50'),'Product_Category_2']=df.loc[(df['Age']=='46-50'),'Product_Category_2'].fillna(int(Product_Category_2_mean.loc['46-50']))
df.loc[(df['Age']=='51-55'),'Product_Category_2']=df.loc[(df['Age']=='51-55'),'Product_Category_2'].fillna(int(Product_Category_2_mean.loc['51-55']))
df.loc[(df['Age']=='55+'),'Product_Category_2']=df.loc[(df['Age']=='55+'),'Product_Category_2'].fillna(int(Product_Category_2_mean.loc['55+']))
df
```

```python
df.loc[(df['Age']=='0-17'),'Lower_Age'] = int(0)
df.loc[(df['Age']=='0-17'),'Upper_Age'] = int(17)
df.loc[(df['Age']=='18-25'),'Lower_Age']= int(18)
df.loc[(df['Age']=='18-25'),'Upper_Age']= int(25)
df.loc[(df['Age']=='26-35'),'Lower_Age']= int(26)
df.loc[(df['Age']=='26-35'),'Upper_Age']= int(35)
df.loc[(df['Age']=='36-45'),'Lower_Age']= int(36)
df.loc[(df['Age']=='36-45'),'Upper_Age']= int(45)
df.loc[(df['Age']=='46-50'),'Lower_Age']= int(46)
df.loc[(df['Age']=='46-50'),'Upper_Age']= int(50)
df.loc[(df['Age']=='51-55'),'Lower_Age']= int(51)
df.loc[(df['Age']=='51-55'),'Upper_Age']= int(55)
df.loc[(df['Age']=='55+'),'Lower_Age']= int(60)
df.loc[(df['Age']=='55+'),'Upper_Age']= int(60)

```

```python
# Removing User Id and product ID column
df.drop('User_ID',  axis =1, inplace = True)
df.drop('Product_ID',  axis =1, inplace = True)
df.drop('Age',  axis =1, inplace = True)
df
```

```python
df['Gender']= df['Gender'].replace("M",1)
df['Gender'] = df['Gender'].replace("F",0)
df['City_Category'] = df['City_Category'].replace("A",0)
df['City_Category'] = df['City_Category'].replace("B",1)
df['City_Category'] = df['City_Category'].replace("C",2)
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].replace('4+', 7) 
```

```python
df.head()
```

```python
features = ['Gender', 'Occupation', 'City_Category', 
       'Stay_In_Current_City_Years', 'Marital_Status', 'Product_Category_1',
       'Product_Category_2', 'Lower_Age', 'Upper_Age']
target = ['Purchase']
```

```python
X = df[features]
y = df[target]
from sklearn.ensemble import RandomForestRegressor
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
X_train = preprocessing.scale(X_train)
X_test = preprocessing.scale(X_test)

```

Important Link For practice and interview questions-

https://stats.stackexchange.com/questions/36165/does-the-optimal-number-of-trees-in-a-random-forest-depend-on-the-number-of-pred

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import accuracy_score
model = RandomForestRegressor(n_estimators =50, random_state = 0)
model.fit(X_train,y_train)
y_predict = model.predict(X_test)
accuracy = model.score(X_test,y_test)
print(accuracy*100)
```
