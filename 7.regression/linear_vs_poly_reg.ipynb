```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Position_Salaries.csv')
df.head()
```

Here in this data we do not need "Position" parameter as this parameter is same as "Level" parameter

```python
X = df.iloc[:,1:2].values
y = df.iloc[:,2].values
```

Here we do not need to split this into train and test data set as we are just analyzing how our model fits the training set points. We also do not need to do Feature scaling here, our machine learning library will automatically handle it and also we do not need to do encoding here because there are no categorical features here. 

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.predict(6.5)
```

```python
#For polynomial regression we need to transform the x and feed it into linear regression model.
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
lin_reg_2.predict(poly_reg.fit_transform(6.5))
```

```python
#Linear regression curve-
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
#Polynomial Regression curve woth degree 2
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
#Polynomial Regression curve woth degree 3
poly_reg = PolynomialFeatures(degree = 3)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly, y)
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_3.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
#Polynomial Regression curve woth degree 4
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg_4 = LinearRegression()
lin_reg_4.fit(X_poly, y)
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_4.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
#Now lets predict the new result using these models.
print("Salary at positional level 6.5 (Linear Regression)",lin_reg.predict(6.5))
print("Salary at positional level 6.5 (Polynomial Regression degree 4)",lin_reg_4.predict(poly_reg.fit_transform(6.5)))
```
